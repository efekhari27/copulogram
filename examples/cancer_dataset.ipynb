```python
import pandas as pd
import seaborn as sns
import copulogram as cp
from matplotlib.colors import to_rgba
```

https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data/data

```python
data = pd.read_csv('data/breast_cancer.csv')
data = data.drop(columns=['id', 'Unnamed: 32'])
data.columns
```

```python
data = data.loc[:, ['diagnosis', 'radius_mean', 'concave points_mean', 'concave points_worst', 'symmetry_mean', 'fractal_dimension_mean']]
data.head()
```

```python
color_palette = sns.color_palette([to_rgba('C3', 0.6), to_rgba('C2', 0.5)], as_cmap=True)
copulogram = cp.Copulogram(data)
copulogram.draw(hue='diagnosis', hue_colorbar=color_palette);
```

```python
copulogram.draw(hue='diagnosis', quantile_contour_levels=[0.1, 0.2, 0.4, 0.6, 0.8, 0.9], kde_on_marginals=True, hue_colorbar=color_palette);
```
